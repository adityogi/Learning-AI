## 1. Series

A **Series** is a one-dimensional, labelled array that can hold any data type.
Think of it as a single column in a spreadsheet — each value has a row label called an **index**.

```
index   value
  0  ->  100
  1  ->  102
  2  ->  104
```

### Creating a Series from different data types

You can create a Series from a list of integers, floats, strings, or booleans.

````python
# Integer Series
int_series = pd.Series([100, 102, 104])
print("Integer Series:")
print(int_series)
print()

# Float Series
float_series = pd.Series([100.1, 102.2, 104.3])
print("Float Series:")
print(float_series)
print()

# String Series
str_series = pd.Series(["A", "B", "C"])
print("String Series:")
print(str_series)
print()

# Boolean Series
bool_series = pd.Series([True, False, True])
print("Boolean Series:")
print(bool_series)
````

### Custom index labels

By default, Pandas assigns integer indices (0, 1, 2, ...).
You can set your own labels using the `index=` parameter.

````python
data = [100, 102, 104]
series = pd.Series(data, index=["a", "b", "c"])
print("Series with custom index:")
print(series)
````

### Accessing elements — `.loc` and `.iloc`

| Method | Access by | Example |
|--------|-----------|---------|
| `.loc` | **Label** (the index name) | `series.loc["a"]` |
| `.iloc` | **Integer position** (0, 1, 2, ...) | `series.iloc[0]` |

Use `.loc` when you have meaningful labels like names or dates.
Use `.iloc` when you just want the 1st, 2nd, 3rd element by position.

````python
series = pd.Series([100, 102, 104], index=["a", "b", "c"])

# .loc -- access by label
print(".loc['a']:", series.loc["a"])    # 100

# .iloc -- access by integer position (0 = first element)
print(".iloc[0]:", series.iloc[0])      # 100  (same element, different syntax)

# Modify a value by label
series.loc["c"] = 200
print("\nAfter setting 'c' to 200:")
print(series)
````

### Filtering a Series

Just like NumPy, you can filter a Series by passing a condition inside `[]`.

````python
my_series = pd.Series([100, 102, 104, 200, 202],
                      index=["a", "b", "c", "d", "e"])

print("Values >= 200:")
print(my_series[my_series >= 200])

print("\nValues < 200:")
print(my_series[my_series < 200])
````

### Creating a Series from a dictionary

When you pass a dictionary, the keys become the index labels automatically.

````python
# Dictionary keys become the index labels
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
series = pd.Series(calories)
print("Calorie tracker:")
print(series)

# Update a value using +=
series.loc["Day 3"] += 500
print("\nAfter adding 500 calories to Day 3:")
print(series)

print("\nDays with 2000+ calories:")
print(series[series >= 2000])

print("\nDays under 2000 calories:")
print(series[series < 2000])
````

## 2. DataFrames

A **DataFrame** is a 2D table with labelled rows and columns — like an Excel spreadsheet.
Each column is a Series, and all columns share the same row index.

```
            Name  Age
Employee 1  Spongebob   30
Employee 2  Patrick     35
Employee 3  Squidward   50
```

### Creating a DataFrame from a dictionary

Keys become column names; the lists become column data.

````python
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age":  [30, 35, 50]
}

df = pd.DataFrame(data)
print(df)
````

### Custom row index

You can set meaningful row labels with `index=` — useful for employee IDs, dates, names, etc.

````python
df = pd.DataFrame(
    {"Name": ["Spongebob", "Patrick", "Squidward"],
     "Age":  [30, 35, 50]},
    index=["Employee 1", "Employee 2", "Employee 3"]
)
print(df)
````

### Accessing rows — `.loc` and `.iloc`

````python
# .loc -- access row by its label
print("Employee 1 (by label):")
print(df.loc["Employee 1"])

print()

# .iloc -- access row by its integer position (0 = first row)
print("Row at position 2 (by integer):")
print(df.iloc[2])
````

### Adding columns

Assign a new list to `df["NewColumnName"]`.

````python
df["Job"] = ["Cook", "N/A", "Cashier"]
print(df)
````

### Adding rows

`df.append()` was removed in newer Pandas versions.
The correct way is to build a small DataFrame for the new row(s) and use `pd.concat()`.

````python
# Adding ONE new row
new_row = pd.DataFrame(
    [{"Name": "Ron", "Age": 25, "Job": "Engineer"}],
    index=["Employee 4"]
)
df = pd.concat([df, new_row])
print(df)
````

````python
# Adding MULTIPLE rows at once -- same idea, just more entries in the list
more_rows = pd.DataFrame(
    [{"Name": "Sandy",  "Age": 28, "Job": "Engineer"},
     {"Name": "Eugene", "Age": 60, "Job": "Manager"}],
    index=["Employee 5", "Employee 6"]
)
df = pd.concat([df, more_rows])
print(df)
````

## 4. Selection — loc & iloc

For the selection examples we use the Pokemon dataset with `Name` as the row index.

`index_col="Name"` tells Pandas to use the Name column as the row label instead of the
default integers 0, 1, 2, ... This lets us look up a Pokemon directly by name with `.loc`.

````python
import io
df = pd.read_csv(io.StringIO(csv_text), index_col="Name")
print(df.head(5))
````

### Selecting columns

````python
# Single column -> returns a Series
print("Height column:")
print(df["Height"])
````

````python
# Multiple columns -> pass a list -> returns a DataFrame
print(df[["Height", "Weight"]].to_string())
````

### `.loc` — select by label

```
df.loc[row_label, column_label]
```
- Row label = the index value (here, the Pokemon name)
- Column label = the column name
- Use `:` for a range of labels — **both ends are inclusive** with `.loc`

````python
# Single row by name
print("Charizard:")
print(df.loc["Charizard"])
````

````python
# Single row, specific columns only
print("Charizard -- Height and Weight:")
print(df.loc["Charizard", ["Height", "Weight"]])
````

````python
# A RANGE of rows (both ends inclusive with .loc)
print("Charizard through Blastoise, Height and Weight:")
print(df.loc["Charizard":"Blastoise", ["Height", "Weight"]])
````

### `.iloc` — select by integer position

```
df.iloc[row_number, column_number]
```
- Uses 0-based integer positions, exactly like NumPy
- The **stop** in a slice is **exclusive** (same as Python list slicing)

````python
# Rows 0-10 (positions 0 through 10)
print("Rows 0-10:")
print(df.iloc[0:11])
````

````python
# Every other row (step = 2)
print("Every 2nd row from positions 0-10:")
print(df.iloc[0:11:2])
````

````python
# Rows 0-10, first 3 columns only
print("Rows 0-10, first 3 columns:")
print(df.iloc[0:11:2, 0:3])
````

### Interactive lookup with error handling

Type a Pokemon name. The `try/except` block handles the case where the name is not in the dataset.

````python
pokemon = input("Enter a Pokemon name (e.g. Pikachu): ")
try:
    print(df.loc[pokemon])
except KeyError:
    print(f"'{pokemon}' was not found in the dataset.")
    print("Available Pokemon:", list(df.index))
````

## 5. Filtering

Filtering means **keeping only the rows that match a condition**.
It works exactly like NumPy boolean indexing:

1. Write a condition -> Pandas creates a True/False mask
2. Pass the mask into `df[...]` -> you get back only the True rows

### Combining conditions — `&`, `|`, `~`

| Operator | Meaning | Example |
|----------|---------|---------|
| `&` | AND — both conditions must be True | `(A) & (B)` |
| `\|` | OR — at least one condition must be True | `(A) \| (B)` |
| `~` | NOT — inverts the condition | `~(A)` |

> **Always wrap each condition in parentheses** when combining them.
> Without parentheses, Python's operator precedence will misparse the expression and give wrong results.

````python
# Reload without Name as index so filtering output is easier to read
df = pd.read_csv(io.StringIO(csv_text))
df.head(3)
````

### Single condition filters

````python
# Pokemon taller than 2 metres
tall = df[df["Height"] >= 2]
print("Tall Pokemon (height >= 2m):")
print(tall[["Name", "Type1", "Height"]].to_string())
````

````python
# Pokemon heavier than 100 kg
heavy = df[df["Weight"] > 100]
print("Heavy Pokemon (weight > 100 kg):")
print(heavy[["Name", "Type1", "Weight"]].to_string())
````

````python
# Legendary Pokemon  (Legendary == 1  or  == True -- both work)
legendary = df[df["Legendary"] == 1]
print("Legendary Pokemon:")
print(legendary[["Name", "Type1", "Type2"]].to_string())
````

````python
# Pokemon whose PRIMARY type is Water
water_primary = df[df["Type1"] == "Water"]
print("Primary Water-types:")
print(water_primary[["Name", "Type1", "Type2"]].to_string())
````

### Combining conditions with `|` (OR) and `&` (AND)

````python
# Water-type in EITHER Type1 OR Type2  ->  use the | (OR) operator
water_any = df[(df["Type1"] == "Water") | (df["Type2"] == "Water")]
print("Any Water-type (primary or secondary):")
print(water_any[["Name", "Type1", "Type2"]].to_string())
````

````python
# Fire primary  ->  single condition
fire = df[df["Type1"] == "Fire"]
print("Primary Fire-types:")
print(fire[["Name", "Type1", "Type2"]].to_string())
````

````python
# Fire primary AND Flying secondary  ->  use the & (AND) operator
fire_flying = df[(df["Type1"] == "Fire") & (df["Type2"] == "Flying")]
print("Fire/Flying dual-types:")
print(fire_flying[["Name", "Type1", "Type2"]].to_string())
````

### Quick reference: `&` vs `|`

```python
# AND -- row must satisfy BOTH conditions
df[(condition_A) & (condition_B)]

# OR -- row needs to satisfy AT LEAST ONE condition
df[(condition_A) | (condition_B)]

# NOT -- invert the condition (keep rows where it is False)
df[~(condition_A)]
```

## 6. Aggregate Functions & GroupBy

**Aggregate functions** reduce a column (or whole DataFrame) to a single summary value.

Common ones: `mean()`, `sum()`, `min()`, `max()`, `count()`

`count()` ignores `NaN` (missing) values. All other functions here also skip NaN by default.

### Whole-DataFrame aggregates

````python
df = pd.read_csv(io.StringIO(csv_text))

# numeric_only=True skips text columns (mean/sum cannot be applied to strings)
print("Mean of each numeric column:")
print(df.mean(numeric_only=True).round(2))
````

````python
print("Sum of each numeric column:")
print(df.sum(numeric_only=True).round(2))
````

````python
print("Min of each numeric column:")
print(df.min(numeric_only=True))
````

````python
print("Max of each numeric column:")
print(df.max(numeric_only=True))
````

````python
# count() shows non-null values per column -- great for spotting missing data
print("Non-null count per column:")
print(df.count())
````

### Single-column aggregates

````python
print("Height -- mean:  ", df["Height"].mean().round(2))
print("Height -- sum:   ", df["Height"].sum().round(2))
print("Height -- min:   ", df["Height"].min())
print("Height -- max:   ", df["Height"].max())
print("Height -- count: ", df["Height"].count(), " (non-null rows only)")
````

### `groupby()` — aggregate per category

`groupby()` splits the DataFrame into groups by a column's unique values, then lets you apply an
aggregate function to each group separately.

**Think of it as:** *"For each Type1, what is the average Height?"*

````python
# Group all rows by their Type1 value
group = df.groupby("Type1")

print("Mean height per Type1:")
print(group["Height"].mean().round(2))
````

````python
print("Total weight per Type1:")
print(group["Weight"].sum().round(1))
````

````python
print("Shortest Pokemon per Type1:")
print(group["Height"].min())
````

````python
print("Tallest Pokemon per Type1:")
print(group["Height"].max())
````

````python
print("Number of Pokemon per Type1:")
print(group["Height"].count())
````

### Multiple aggregates at once with `.agg()`

Instead of calling one function at a time, you can get several statistics in one call.

````python
summary = group["Height"].agg(["mean", "min", "max", "count"])
print("Height stats per Type1:")
print(summary.round(2))
````

## 7. Data Cleaning

Data cleaning is the process of fixing or removing **incomplete, incorrect, or irrelevant data**.

> In real projects, roughly 75% of time spent with Pandas is on data cleaning — so this is one of the most important sections!

### The main tasks
1. Drop irrelevant columns
2. Handle missing values (`NaN`)
3. Fix inconsistent values
4. Standardise text
5. Fix data types
6. Remove duplicate rows

````python
df = pd.read_csv(io.StringIO(csv_text))
print("Columns:", df.columns.tolist())
print("Shape:", df.shape)
````

### 1. Drop irrelevant columns

`df.drop(columns=[...])` returns a **new** DataFrame — the original is not modified unless you
reassign the result: `df = df.drop(...)`.

````python
# Drop the Legendary column
df_clean = df.drop(columns=["Legendary"])
print("Columns after dropping Legendary:", df_clean.columns.tolist())
````

````python
# Drop multiple columns at once
df_clean = df.drop(columns=["Legendary", "No"])
print(df_clean.head(3))
````

### 2. Handle missing values

`NaN` (Not a Number) is Pandas' placeholder for a missing value.

| Method | What it does |
|--------|-------------|
| `df.isnull().sum()` | Count missing values per column |
| `df.dropna(subset=[...])` | Remove rows that have NaN in those columns |
| `df.fillna({col: val})` | Fill NaN in the specified column with a value |

````python
print("Missing values per column:")
print(df.isnull().sum())
````

````python
# Drop rows where Type2 is missing (many Pokemon have only one type)
df_no_missing = df.dropna(subset=["Type2"])
print(f"Rows before: {len(df)}  |  After dropping missing Type2: {len(df_no_missing)}")
print(df_no_missing[["Name", "Type1", "Type2"]].to_string())
````

````python
# Fill missing Type2 with the string "None" instead of removing the row
df_filled = df.fillna({"Type2": "None"})
print("After filling missing Type2 with 'None':")
print(df_filled[["Name", "Type1", "Type2"]].to_string())
````

### 3. Fix inconsistent values

Use `.replace()` to correct typos, standardise capitalisation, or remap categories.

````python
df_fixed = df_filled.copy()
df_fixed["Type1"] = df_fixed["Type1"].replace({
    "Grass": "GRASS",
    "Fire":  "FIRE",
    "Water": "WATER"
})
print(df_fixed[["Name", "Type1"]].to_string())
````

### 4. Standardise text

The `.str` accessor gives you string methods that work on an entire column at once.

````python
df_text = df_filled.copy()

# Convert all names to lowercase
df_text["Name"] = df_text["Name"].str.lower()
print(df_text[["Name", "Type1"]].head())

# Other handy .str methods:
#   .str.upper()              -> ALL CAPS
#   .str.title()              -> Title Case
#   .str.strip()              -> remove leading/trailing whitespace
#   .str.replace("old","new") -> find and replace within strings
````

### 5. Fix data types

`df["col"].astype(new_type)` converts a column to a different data type.

Common conversions: `int` -> `bool` (0/1 flags), `object` -> `int` or `float`.

````python
df_types = df.copy()
print("Before -- Legendary dtype:", df_types["Legendary"].dtype)
print(df_types["Legendary"].head())

df_types["Legendary"] = df_types["Legendary"].astype(bool)
print("\nAfter -- Legendary dtype:", df_types["Legendary"].dtype)
print(df_types["Legendary"].head())
````

### 6. Remove duplicate rows

`drop_duplicates()` removes rows that are identical across all (or specified) columns.

````python
# Create a small DataFrame with deliberate duplicates to demonstrate
dup_data = pd.DataFrame({
    "Name":  ["Pikachu", "Squirtle", "Pikachu", "Charmander", "Squirtle"],
    "Type1": ["Electric", "Water", "Electric", "Fire", "Water"],
    "HP":    [35, 44, 35, 39, 44]
})
print("With duplicates:")
print(dup_data)

# Remove exact duplicate rows (keeps the first occurrence by default)
deduped = dup_data.drop_duplicates()
print("\nAfter removing duplicates:")
print(deduped)
````

````python
# Deduplicate based on specific columns only
# keep="last" keeps the last occurrence instead of the first
deduped_by_name = dup_data.drop_duplicates(subset=["Name"], keep="last")
print("Deduplicated on Name only (keeping the last occurrence):")
print(deduped_by_name)
````

## Pandas Cheat Sheet

### Series
| Task | Code |
|------|------|
| Create from list | `pd.Series([1, 2, 3])` |
| Custom index | `pd.Series(data, index=["a","b","c"])` |
| Access by label | `s.loc["a"]` |
| Access by position | `s.iloc[0]` |
| Filter | `s[s > 100]` |

### DataFrame
| Task | Code |
|------|------|
| Create | `pd.DataFrame({"Col": [...]})` |
| Read CSV | `pd.read_csv("file.csv")` |
| Read JSON | `pd.read_json("file.json")` |
| First N rows | `df.head(5)` |
| Column types | `df.dtypes` |
| Summary stats | `df.describe()` |
| Select column | `df["Col"]` |
| Select columns | `df[["A","B"]]` |
| Row by label | `df.loc["row_label"]` |
| Row by position | `df.iloc[0]` |
| Filter rows | `df[df["Col"] > 5]` |
| AND filter | `df[(A) & (B)]` |
| OR filter | `df[(A) \| (B)]` |
| Add column | `df["New"] = [...]` |
| Add rows | `pd.concat([df, new_rows])` |

### Aggregation
| Task | Code |
|------|------|
| Column mean | `df["Col"].mean()` |
| Column sum | `df["Col"].sum()` |
| Group & aggregate | `df.groupby("Col")["Other"].mean()` |
| Multiple stats | `df.groupby("Col")["Other"].agg(["mean","min","max"])` |

### Data Cleaning
| Task | Code |
|------|------|
| Count missing | `df.isnull().sum()` |
| Drop rows with NaN | `df.dropna(subset=["Col"])` |
| Fill NaN | `df.fillna({"Col": "value"})` |
| Drop columns | `df.drop(columns=["Col"])` |
| Replace values | `df["Col"].replace({"old": "new"})` |
| Lowercase text | `df["Col"].str.lower()` |
| Change type | `df["Col"].astype(bool)` |
| Remove duplicates | `df.drop_duplicates()` |

## Reshapping Arrays

```python
# Creating array of 10 Elements, aka a list with 10 elements
myarray = np.arange(10)
print(f"The generated array is: {myarray}")
# Reshaping this array into a 5 * 2 Array, i.e. into an 5 X 2 matrix.
new_matrix = myarray.reshape(5, 2)
print(f"\nThe new array is: \n{new_matrix}")
```

```python
# Flattening of Arrays or Networks
flattened = new_matrix.flatten() # Converts an n-dimensional array into a single dimensional array.
print(f"The flattened array/vector/matrix is: \n{flattened}")
flattened.shape
```

```python
# Creating a 2 X 3 Array
t_array = np.array([[1, 2, 3], [4, 5, 6]]) # This is a 2 Dimensional Matrix
print(f"The created array is: \n{t_array}")
# Finding the transpose of the array generated
# Transpose - Replaces rows with columns.
transpose_a = t_array.transpose() # When dealing with 3 Dimensional Matrices we can pass additional parameters inside the Transpose Function
# We can alternate the axis
print(f"\nThe transpose of the array created is: \n{transpose_a}")
```

```python
# Creating a 2 X 3 Array
t_array = np.zeros((3, 3, 3)) # 3 Dimensional Matrix, it has rows, columns and depth
t_array[0, 0, 1] = 1
t_array[0, 1, 0] = 5

print(f"The created array is: \n{t_array}")
# Finding the transpose of the array generated
# Transpose - Replaces rows with columns.
# Transpose converts t_array[i, j, k] to ====>  t_array[i, k, j] so transpose_a[0, 1, 0] = 1 and transpose_a[0, 0, 1] = 5
transpose_a = t_array.transpose((0, 2, 1)) # When dealing with 3 Dimensional Matrices we can pass additional parameters inside the Transpose Function
# We can alternate the axis
print(f"\nThe transpose of the array created is: \n{transpose_a}")
```

```python
t_array.shape # Zeroth dimension in the zeroth place, 2nd dimension in the 1st place and the 1st dimension in the 2nd place.
# Before Transpose ===>
# t_array[i, j, k] # Can access the elements of t_array using 3 things , i is the 0th dimension, j is the 1st dimension and k is the 3rd dimension.
# After Transpose ===>
# t_array[i, k, j]
```

In higher dimensions transpose can mean many things
2D Arrays are still matrics and as we go to higher dimensions we start calling them tensors at that point.

For Examples:
For a 2D Matrix Transpose just means exchanging the rows an columns but when, let's say a matrix has 3 Dimensions or more say rows, columns and depth the exchange order can be in many different ways.
(0, 1, 2) is the original matrix or tensor order, you can have (0, 2, 1), (1, 2, 0) or (1, 0, 2), etc....

So as we can permute these in many ways so we have various parameters in transpose.

```python
# Creating random arrays.
# Create a random 3 X 2 Array
arrays = np.random.random_sample((3, 2)) * 5 # Create an array with shape i.e. func np.random.random_sample((3, 2)) and then multiply each element in the array with 5
# Random func gives random values b/w 0 and 1 and then for showing purposes we have multiplied it with 5
print(f"The array generated is: \n{arrays}")
# Way to fix randomness, is by using the value of seed
# The seed function is still somewhat deterministic
np.random.seed(4) # If you don't fix the seed value then each time it takes the random value for the seed.
# The randomness behind these programs is something called as sudo randomness, i.e., there is some order behind this randomness, feeding in some number suggests that the random logic will start at that particular number, this is important when you want predictable numbers in matrix, this is for the value of seed.
# Fixing the value of seed always gives us the same set of random values
print(f"The array generated is: \n{arrays}")
```

```python
# Finding the index of the minimum element in an array
print("Original Array: \n", arrays)
# Finding the minimum element in an entire array.
print(f"\nThe minimum value in the entire array is: \nmin(array) = {np.min(arrays, 1)}") # 0 denotes row and 1 denotes column
# Finding the index of the minimum value
print(f"\nThe index of the minimum value of the array is: \nargmin(arrays) = {np.argmin(arrays, 1)}")
```

```python
# Finding the index of the Maximum element in an array
print("Original Array: \n", arrays)
# Finding the Maximum element in an entire array.
print(f"\nThe maximum value in the entire array is: \nmax(array) = {np.max(arrays)}") # If we specify axis then it gives the max/min in that axis
# Finding the index of the Maximum value
# np.argmax(a, axis = None, out = None) Returns maximum value from the row or take the maximum value from the column
print(f"\nThe index of the maximum value of the array is: \nargmax(arrays) = {np.argmax(arrays)}")
```

```python
# Find the maximums and minimums of Rows 1 and 2
# np.maximum(x1, x2) is same as np.where(x1 >= x2, x1, x2)
print(f"Maximum of row1, row2 = {np.maximum(arrays[1], arrays[2])}")
print(f"Minimum of row1, row2 = {np.minimum(arrays[1], arrays[2])}")
```

## Linear Algebra

```python
# Create a random square matrix say a
a = np.random.random_sample((3, 3)) # Random 3 X 3 Matrix
b = np.exp(a) # Finding the exponential of Matrix a
print(f"b = exp(a){b}")
```

Difference b/w a 10 X 1 matrix (2D Matrix, with 2nd dimension being of size 1) and a 10 X - matrix(1D Tensor where 2nd dimension dosnt even exisit), its the same meaning for humans wheras numpy treats it as different matrixes

```python
# Compute the Determinent, Rank, and Trace of b
det = np.linalg.det(b) #Linalig pertains to all the functions in NumPY for Linear Algebra
print(f"determinant(b) = {det}")
rank = np.linalg.matrix_rank(b) # To find the rank of the matrix
print(f"rank(b) = {rank}")
trace = np.trace(b) # To find the trace of the matrix
print(f"trace(b) = {trace}")
```

```python
# Find the eigen values and the eigen vectors of square matrix b
eigen_val, eigen_vec = np.linalg.eig(b)
print(f"The eigen values of b: \n{eigen_val}")
print(f"The right eigen vectors of b: \n{eigen_vec}")
```

$Av = \lambda v$ lambda is called eigen value

```python
# Implement singular value decomposition (svd) on matrix b
# Eigen Value Decomposition is sort of related to SVD.
u, s, vh = np.linalg.svd(b) # Resturns 3 vectors u, s and vh
print(f"U of b is: \n{u}")
print(f"V of b is: \n{vh}")
print(f"singular values of b are: \n{s}")
```

## Combining overlapping Data Sets

```python
import numpy as np
```

```python
df1 = pd.DataFrame({'A': [1., np.nan, 3., 5., np.nan],
                    'B': [np.nan, 2., 3., np.nan, 6.]})
df2 = pd.DataFrame({'A': [5., 2., 4., np.nan, 3., 7.],
                    'B': [np.nan, np.nan, 3., 4., 6., 8.]})
```

```python
df1
```

```python
df2
```

```python
df1.combine_first(df2)
```

**combine_first() combines two DataFrame objects where missing values in one DataFrame are conditionally filled with like-labelled values from the other DataFrame.**

```python
def combiner(x, y):
  return np.where(pd.isna(x), y, x)
```

```python
df1 = pd.DataFrame({'A': [1., 2., 3., 5., 6.],
                    'B': [1., np.nan, 3., 4., np.nan]})
df2 = pd.DataFrame({'A': [5., 2., 4., 2., 3.],
                    'B': [7., 3., 4., 6., 8.]})

df3 = combiner(df1, df2)
df3
```

**To exclude missing data you can use skipna = True**

```python
df1.sum(axis = 1, skipna = True)
```

## Crosstabs

```python
raw_data = {
    'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks',
                 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons',
                 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
    'company': ['infantry', 'infantry', 'cavalry', 'cavalry',
                'infantry', 'infantry', 'cavalry', 'cavalry',
                'infantry', 'infantry', 'cavalry', 'cavalry'],
    'experience': ['veteran', 'rookie', 'veteran', 'rookie',
                   'veteran', 'rookie', 'veteran', 'rookie',
                   'veteran', 'rookie', 'veteran', 'rookie'],
    'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon',
             'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
    'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]
}

df = pd.DataFrame(raw_data)
```

## Counting the number of observations by regiment and category

```python
pd.crosstab(df.regiment, df.company, margins = True)
```

## Create a crosstab of the number of rookie and veteran cavalry and infantry soldiers per regiment.

```python
pd.crosstab([df.company, df.experience], df.regiment, margins = True)
```

## Series
A series is a pandas 1-Dimensional labelled array that can hold any data type.
Think of it like a single columnn in a spreadsheet (1-Dimensional)

```python
data = [100, 102, 104]
series = pd.Series(data) # Series Consructor
print(series)
```

```python
data = [100,1, 102.2, 104.3]
series = pd.Series(data) # Series Consructor
print(series)
data = ["A", "B", "C"]
series = pd.Series(data) # Series Consructor
print(series)
data = [True, False, True]
series = pd.Series(data) # Series Consructor
print(series)
```

```python
data = [100, 102, 104]
series = pd.Series(data, index = ["a", "b", "c"]) # Series Consructor
print(series)
print(series.loc["a"]) # Location by Label
# print(series.loc["d"]) # Key Error
series.loc["c"] = 200
print(series)
print(series.iloc[0]) # Integer Location
```

```python
my_data = [100, 102, 104, 200, 202]
series = pd.Series(my_data, index = ["a", "b", "c", "d", "e"])
print(series[series >= 200])
print(series[series < 200])
```

```python
calories = {"Day 1": 1750, "Day2": 2100, "Day3": 1700}

series = pd.Series(calories)
print(series)
series.loc["Day3"] += 500
print(series.loc["Day2"])
print(series.loc["Day3"])
print(series[series >= 2000])
print(series[series < 2000])
print(series)
```

## DataFrames
In pandas a tabular data structure with rows and columns is called a DataFrame.
It is 2-Dimensional and is similar to an Excel Spreadsheet

```python
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}

df = pd.DataFrame(data) #Constructor rather than function
print(df)
```

```python
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}
df = pd.DataFrame(data, index = ["Employee 1", "Employee 2", "Employee 3"])
print(df)
```

```python
print(df.loc["Employee 1"]) #LOC - Location by Label
print("\n")
print(df.iloc[2]) #Select by integer location
```

```python
# Add a new column
df["Job"] = ["Cook", "N/A", "Cashier"]
print(df)
```

```python
# Adding a new row - Easy way is to create a new data frame and then concatenate it.
new_row = pd.DataFrame([{"Name": "Ron", "Age": 25, "Job": "Engineer"}],
                       index = ["Employee 4"]) # Seperate dataframe with a single row
df = pd.concat([df, new_row]) #Append is depricated
print(df)
```

```python
# Adding new rows - Easy way is to create a new data frame and then concatenate it.
new_row = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Engineer"},
                        {"Name": "Eugene", "Age": 60, "Job": "Manager"}],
                       index = ["Employee 5", "Employee 6"]) # Seperate dataframe with a single row
df = pd.concat([df, new_row]) #Append is depricated
print(df)
```

## Filtering
Keeping the rows that match a condition when working with a data frame

```python
heavy_pokemon = df[df["Weight"] > 100]
print(heavy_pokemon)
```

```python
legendary_pokemon = df[df["Legendary"] == 1]
# legendary_pokemon = df[df["Legendary"] == True]
print(legendary_pokemon)
```

```python
water_pokemon = df[df["Type1"] == "Water"]
print(water_pokemon)
```

```python
water_pokemon = df[(df["Type1"] == "Water") |
                   (df["Type2"] == "Water")] # Revisit and and & difference and soforth. # Seastyle Operators
print(water_pokemon)
```

```python
ff_pokemon = df[df["Type1"] == "Fire"]
print(ff_pokemon)
```

```python
ff_pokemon = df[(df["Type1"] == "Fire") & (df["Type2"] == "Flying")]
print(ff_pokemon)
```

## Aggregate Functions OR Aggregation
Aggregate Functions reduce a set of values into a single summary value.
It is used to summarize and analyze data.
It is often used with the groupby() function.

```python
df = pd.read_csv("data.csv")
```

### Whole DataFrame Aggregate Functions

```python
# print(df.mean())
# Will result in error as it is. Mean function cannot be applied on anything that is non-numeric
print(df.mean(numeric_only = True))
```

```python
print(df.sum(numeric_only = True))
```

```python
print(df.min(numeric_only = True))
```

```python
print(df.max(numeric_only = True))
```

```python
print(df.count()) #Count dosn't include any null values
```

```python
print(df["Height"].min())
```

```python
print(df["Height"].max())
```

## Data Cleaning using Pandas
It is the process of fixing/removing:
- Incomplete
- Incorrect
- and Irrelevent Data

~75% of the work done with pandas is around data cleaning.

```python
# Drop irrelevant columns
new_df = df.drop(columns = ["Legendary"]) # Retuns a new dataframe with the deleted column
print(new_df)
```

```python
print(df)
```

```python
# Drop irrelevant columns
new_df = df.drop(columns = ["Legendary", "No"]) # Retuns a new dataframe with the deleted column
print(new_df)
```

```python
# Handle Missing Data
new_df = df.dropna(subset = ["Type2"]) #Dropna = Drop Not Available
print(new_df.to_string())
print(new_df.count())
```

```python
new_df = df.fillna({"Type2": "None"})
print(new_df.to_string())
```

```python
# Fix Inconsistent Values
new_df["Type1"] = df["Type1"].replace({"Grass": "GRASS",
                                       "Fire": "FIRE",
                                       "Water": "WATER"})
print(new_df.to_string())
```

```python
# Standardize Text
new_df["Name"] = new_df["Name"].str.lower()
print(new_df.to_string())
```

```python
# Fix Data Types
df["Legendary"] = df["Legendary"].astype(bool)
print(df.to_string())
```

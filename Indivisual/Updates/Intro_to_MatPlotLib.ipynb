```python
import numpy as np

x = np.array([2023, 2024, 2025, 2026]) # List of x-coordinates
y = np.array([15, 25, 30, 20]) # List of y-coordinater

plt.plot(x, y, marker = ".",
         markersize = 30,
         markerfacecolor = "cyan",
         markeredgecolor = "#1cd3fc",
         linestyle = "dotted",
         linewidth  = 4,
         color = "#1c5bfc") # (Coordinates on x-axis, Coordinates of y-axis)
         # mfc = "#1cd3fc" <---- This is makerfacecolor
plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026]) # List of x-coordinates
y1 = np.array([15, 25, 30, 20]) # List of y-coordinater
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])

line_style = dict(marker = ".",
         markersize = 30,
         markerfacecolor = "cyan",
         markeredgecolor = "#1cd3fc",
         linestyle = "dotted",
         linewidth  = 4,
         color = "#1c5bfc")

plt.plot(x, y1, **line_style)
plt.plot(x, y2)
plt.plot(x, y3, **line_style)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])

plt.title("Class Size", fontsize = 20, family = "Arial", fontweight = "bold", color = "#2d4cfc")

plt.xlabel("Year", fontsize = 20, family = "Arial", fontweight = "bold", color = "#2dbefc")

plt.ylabel("Students", fontsize = 20, family = "Arial", fontweight = "bold", color = "#2dbefc")

plt.tick_params(axis = "both", colors = "#2dbefc")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x)

plt.show()
```

## grid()
Helps make plots easier to read by adding reference lines.


```python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.grid()

plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.grid(axis = "x", linewidth = 2, color = "lightgrey", linestyle = "dashdot")
# plt.grid(axis = "y")

plt.plot(x, y)
plt.show()
```

## Bar Charts
Compare categories of data by representing each category with a bar.

```python
categories = np.array(["Grains", "Fruit", "Vegetables", "Protein", "Dairy", "Sweets"])
values = np.array([4, 3, 2, 5, 3, 1])

plt.bar(categories, values, color = "skyblue")

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

## Pie Charts
They are circular charts divided into slices to show percentages of the total.
Good for vizualizing distribution among categories.

```python
categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])
colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels = categories, autopct = "%1.1f%%", colors = colors, explode = [0, 0, 0, 0.1], shadow = True, startangle = 90)

plt.title("Pie Chart")

plt.show()
```

## Scatter Graph
Shows the relationship between two variables.
Helps to identify a correlation (+, -, None)
Example: Study hours vs. text scores

```python
x1 = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] # Hours studied
y1 = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87] # Grades

x2 = np.array([0, 1, 2, 2, 3, 4, 5, 6, 7, 8, 8]) # Hours Studied
y2 = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87]) # Grades

plt.scatter(x1, y1, color = "skyblue", alpha = 0.5, s = 200, label = "Class A")

plt.scatter(x2, y2, color = "skyblue", alpha = 0.5, s = 200, label = "Class B")

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grades")

plt.legend()
plt.show()
```

## Histogram
A visual representation of the distribution of qualitative data.
They group values into bins (intervals) and counts how many fall in each range.

```python
scores = np.random.normal(loc = 80, scale = 10, size = 100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins = 10, color = "lightgreen", edgecolor = "black")

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("# of students")

plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])

figure, axes = plt.subplots(2, 2) # axes is technically a NumPY array
# figure, axes = plt.subplots(2, 1)
# figure, axes = plt.subplots(3, 2)

#axes[0, 0].plot()

#plt.show()

axes[0, 0].plot(x, x*2, color = "red")
axes[0, 0].set_title("x*2")

axes[0, 1].plot(x, x**2, color = "blue")
axes[0, 1].set_title("x**2")

axes[1, 0].plot(x, x**3, color = "green")
axes[1, 0].set_title("x**3")

axes[1, 1].plot(x, x**4, color = "purple")
axes[1, 1].set_title("x**4")

plt.show()
```
